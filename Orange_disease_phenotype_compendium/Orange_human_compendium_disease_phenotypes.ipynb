```python
###setup

import urllib.request
import json 
import datetime
import pandas as pd
import numpy as np
import array
import matplotlib
import seaborn
```

```python
print(datetime.datetime.now().time())

disease_phenotype_data = []
###get human phenotype data
#curl -X GET --header 'Accept: application/json' 'https://api.monarchinitiative.org/api/bioentity/disease/OMIM%3A605543/phenotypes/?rows=20&fetch_objects=true'gene_data = []
with urllib.request.urlopen("https://api.monarchinitiative.org/api/mart/disease/phenotype/NCBITaxon%3A9606") as url:
    disease_phenotype_data.append(json.loads(url.read().decode()))

print(datetime.datetime.now().time())
```

```python
disease_phenotype_data[0][0]
```

```python
dis_phe_dict = dict()
phenotype_dict = dict()

for i in range(len(disease_phenotype_data[0])):
    subj = disease_phenotype_data[0][i]['subject']
    dis_phe_dict[subj] = disease_phenotype_data[0][i]['objects']
    for obj in disease_phenotype_data[0][i]['objects']:
        phenotype_dict[obj] = 1
        
        
print(len(disease_phenotype_data[0]))
print(len(dis_phe_dict))
```

```python

diseases = dis_phe_dict.keys()
phenotypes = phenotype_dict.keys()

dis_phe_mat = np.zeros(shape = (len(diseases), len(phenotypes)))

count = 0
for dis in diseases:
    curphe = dis_phe_dict[dis]
    
    for phe in curphe:
        phe_index = list(phenotypes).index(phe)
        dis_phe_mat[count][phe_index] = 1
    if count % 1000 == 0:
        print (count)
    count = count+1

dis_phe_df = pd.DataFrame(dis_phe_mat,columns=phenotypes, index=diseases)

print(dis_phe_df.shape)

print(datetime.datetime.now().time())
```

```python
###output full annotation matrix to file, for the record
print(datetime.datetime.now().time())
#np.savetxt(r'./human_GO_annotation_matrix.txt', go_annot_df.values, fmt='%d')
dis_phe_df.to_csv(r'human_disease_phenotype_matrix.txt', sep='\t')
print(datetime.datetime.now().time())
```
