```python
###setup

import urllib.request
import json 
import datetime
import pandas as pd
import numpy as np
import array
import matplotlib
import seaborn

from ontobio.ontol_factory import OntologyFactory
```

```python
print(datetime.datetime.now().time())

disease_phenotype_data = []
###get human phenotype data
#curl -X GET --header 'Accept: application/json' 'https://api.monarchinitiative.org/api/bioentity/disease/OMIM%3A605543/phenotypes/?rows=20&fetch_objects=true'gene_data = []
with urllib.request.urlopen("https://api.monarchinitiative.org/api/mart/disease/phenotype/NCBITaxon%3A9606") as url:
    disease_phenotype_data.append(json.loads(url.read().decode()))

print(datetime.datetime.now().time())
```

```python
disease_phenotype_data[0][0]
```

```python
ofactory = OntologyFactory()
hp_onto = ofactory.create("hp") 

#ofactory = OntologyFactory()
mondo_onto = ofactory.create("mondo") 
```

```python
dis_phe_dict = dict()
phenotypes = []#dict()
diseases = []#dict()

for i in range(len(disease_phenotype_data[0])):
    subj = disease_phenotype_data[0][i]['subject']
    diseases.append(subj)
    #print(subj)
    dis_phe_dict[subj] = disease_phenotype_data[0][i]['objects']
    for obj in disease_phenotype_data[0][i]['objects']:
        #phenotype_dict[obj] = 1
        phenotypes.append(obj)
        
        
#print(len(disease_phenotype_data[0]))
#print(len(dis_phe_dict))
```

```python
#dis_phe_dict
#print(phenotype_dict.keys())
#print(hp_onto.label(phenotype_dict.keys()))

phenotype_descr = []
for i in range(0, len(phenotypes)):
    phenotype_descr.append(hp_onto.label(phenotypes[i]))
                           
#print(phenotype_descr)

disease_descr = []
for i in range(0, len(diseases)):
    disease_descr.append(mondo_onto.xrefs(diseases[i]))
    disease_descr.append(mondo_onto.label(diseases[i]))
    
#print(disease_descr)                         
```

```python
pheno_labels_mat = np.zeros(shape = (len(phenotypes), 2))

pheno_labels_df = pd.DataFrame(pheno_labels_mat,columns=["id","label"])
pheno_labels_df['id'] = phenotypes
pheno_labels_df['label'] = phenotype_descr

###output phenotype ids and labels
print(datetime.datetime.now().time())
pheno_labels_df.to_csv(r'phenotype_labels.txt', sep='\t')
print(datetime.datetime.now().time())
```
